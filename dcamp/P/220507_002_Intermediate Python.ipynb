## Matplotlib

```python
# line plot

import matplotlib.pyplot as plt

x = list(range(1, 11, 1))
y = [1, 5, 8, 15, 26, 35, 50, 63, 82, 99]

print(x)
print(y)
print('')

plt.plot(x, y)
plt.show()
```

```python
# scatter plot

import matplotlib.pyplot as plt

x = list(range(1, 11, 1))
y = [1, 5, 8, 15, 26, 35, 50, 63, 82, 99]

print(x)
print(y)
print('')

plt.scatter(x, y)
plt.show()

print('')
print(y[-3:]) # last 3 elements of y
```

```python
# scatterplot with both axes on log scale

import matplotlib.pyplot as plt
from sklearn import datasets

iris_data = datasets.load_iris().data

plt.scatter(iris_data[:, 0], iris_data[:, 1])
plt.xscale('log')
plt.yscale('log')

plt.show()
```

```python
# histograms - varying number of bins

import matplotlib.pyplot as plt
from sklearn import datasets

iris_data = datasets.load_iris().data

plt.hist(iris_data[:, 1])
plt.show()
plt.clf() # clear current figure

plt.hist(iris_data[:, 1], bins = 5)
plt.show()
plt.clf()

plt.hist(iris_data[:, 1], bins = 15)
plt.show()
plt.clf()
```

```python
# label your axes, add a title, modify ticks

import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets

iris = datasets.load_iris()
iris_data = iris.data
iris_labels = iris.target

#print(iris_data)
#print(iris_labels)
#help(iris)

#iris.__getattr__

# how do I get names of all attributes e.g. column names?

# help(plt)
# help(plt.xticks)

plt.scatter(iris_data[:, 0], iris_data[:, 1])

plt.xlabel('this is the x axis')
plt.ylabel('this is the y axis')
plt.title('first two columns of iris')
plt.yticks(ticks = np.arange(1.5, 5.5, 0.5), 
          labels = ['1.5cm', '2.0cm' , '2.5cm', '3.0cm' , '3.5cm', '4.0cm', '4.5cm', '5.0cm'], 
          rotation = 45)
#list(range(1.5, 5, 0.5)) --> doesn't work! range only produces integers; use np.arange


plt.show()
```

```python
# more plotting and adding axis labels and title

import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets

iris = datasets.load_iris()

iris_data = iris.data
iris_labels = iris.target

plt.scatter(iris_data[:, 2], iris_data[:, 3])

xlab = 'iris column 2 [in cm]'
ylab = 'iris column 3 [in cm]'
title = 'iris column 2 by column 3'

plt.xlabel(xlab)
plt.ylabel(ylab)
plt.title(title)

plt.show()
```

```python
# customizing ticks

import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets

iris = datasets.load_iris()
iris_data = iris.data
iris_labels = iris.target

plt.scatter(iris_data[:, 1], iris_data[:, 2])

xlab = 'iris column 1 [in cm]'
ylab = 'iris column 2 [in cm]'
title = 'iris columns 1 and 2'

plt.xlabel(xlab)
plt.ylabel(ylab)
plt.title(title)

xticks_ticks = np.arange(1, 6, 1)
xticks_labels = ['one', 'two', 'three', 'four', 'five']

plt.xticks(ticks = xticks_ticks,
          labels = xticks_labels,
          rotation = -45)

plt.show()
```
